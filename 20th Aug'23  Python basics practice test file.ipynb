```python
#42. Encode a string using a Caesar cipher.
s1 = input('Enter a string: ')
shift= int(input('shift '))
ceaser=""
for char in s1:
  if char.isalpha():
    if char.isupper():
       charnew=chr((ord(char)+shift-65)%26+65)
       ceaser=ceaser+charnew
    if char.islower():
        charnew=chr((ord(char)+shift-97)%26+97)
        ceaser=ceaser+charnew
  else:
     ceaser=ceaser+char
print(ceaser)

```

```python
# 49. Check if a tuple contains another tuple as a subsequence.
t=(1,2,3,4,5,6.7,8,9)
t1=(1,2,3)
if t1[0] in t:
  k=t.index(t1[0])
  for i in range(len(t1)):
    if t[k+i]!=t1[i]:
      print("not subsequence")
      break
  else:
    print("subsequence")
else:
  print("not subsequence")
```
